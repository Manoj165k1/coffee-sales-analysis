```python
import pandas as pd
import numpy as np
a=pd.read_excel("Coffe_sales.xlsx")
a.head()
```

```python
#info function - give information about the dataset
a.info()
```

```python
a['date']=a['date'].astype('datetime64[ns]') #here we change the data tye of date object to datetime type
a.info()
```

```python
a['Day'] = a['date'].dt.day_name()       # Day name (e.g., Monday)
a['Month'] = a['date'].dt.month_name()   # Month name (e.g., July)
a['Day_of_Month'] = a['date'].dt.day     # date
a
```

```python
#GROUPBY
#groupby - particular group- df.groupby(col)
#groups  - df.groupby(col).groups
#get_group - af.groupby(col).get_group("groupname")
#sigle agg  - df.groupby(col)[colname].agg
#multiple agg - df.groupby(col).agg({col:agg,col:agg})
#multiple groups - df.groupby([col1,col2])
```

GROUPBY QUESTIONS:

1.What is the total sales amount for each coffee product?

2.What is the average money per weekday

3.Which hour_of_day generates the highest total sales?

4.What is the average sales (money) for each Month_name?

5.What is the average sales amount (money) for each cash_type (Cash/Card)?

```python
#1.What is the total sales amount for each coffee product?
c=np.round(a.groupby("coffee_name")["money"].sum(),2)
c
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,4))
plt.barh(c.index,c.values)
plt.title("total sales per category")
plt.xlabel("money")
plt.ylabel("coffe name")
plt.grid(axis='x')
plt.show()
```

1.Latte has the highest sales, crossing 27,000

2.Americano with Milk is the second best-seller with sales above 25,000.

3.Cappuccino and Americano have moderate sales 15,000 to 18,000.

4.Hot Chocolate, Cocoa, and Cortado have lower sales, around 7,000 to 10,000.

5.Espresso has the lowest sales below 3,000,

```python
#2.What is the average money per weekday
d=np.round(a.groupby("Weekday")["money"].mean(),1)
d
```

```python
plt.figure(figsize=(8,5))
plt.plot(d.index,d.values,linestyle='-', color="green")
plt.title("Average Sales by Weekday")
plt.xlabel("Weekday")
plt.ylabel("Average Sales (money)")
plt.grid()
plt.show()
```

1.Monday has the highest average sales 32

2.Saturday records the lowest sales 31.5

3.Sunday and Tuesday close to 31.9, almost matching Monday.

4.Friday and Thursday are mid-level as 31.7.

5.Wednesday is lower at 31.6, but not as low as Saturday.

```python
#3.Which hour_of_day generates the highest total sales?
e=np.round(a.groupby("hour_of_day")["money"].sum(),1)
e
```

```python
plt.figure(figsize=(10,6))
plt.bar(e.index,e.values,color="blue",edgecolor="black")
plt.title("Total Sales by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Total Sales")
plt.grid(axis="y")
plt.show()
```

1.Peak sales occur around 10 AM

2.Another high point is at 4 PM (16:00)

3.Sales start rising from 7 AM onwards

4.Lowest sales are early morning (before 7 AM) and late evening (after 9 PM).

5.Sales decrease after 6 PM,

```python
#4.What is the average sales (money) for each Month_name?
f=np.round(a.groupby("Month_name")["money"].mean(),1)
f
```

```python
plt.figure(figsize=(10,6))
plt.plot(f.index,f.values,marker="o",color="green", linewidth=2)
plt.title("Average Sales by Month")
plt.xlabel("Month")
plt.ylabel("Average Sales (money)")
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

1.April and June recorded the highest average sales above 34.

2.August had the lowest average sales around 28.

3.May also showed strong performance, slightly below April and June.

4.September and July were on the lower side compared to other months around 29 to 30.

```python
#5.What is the average sales amount (money) for each cash_type (Cash/Card)?
g=np.round(a.groupby("cash_type")["money"].mean(),1)
g
```

```python
plt.figure(figsize=(8,6))
plt.barh(g.index,g.values,color="blue",edgecolor="black")
plt.title("Total Sales by Hour of Day")
plt.xlabel("money")
plt.ylabel("cash type")
plt.grid(axis="x")
plt.show()
```

Cash transactions generated the highest total sales around 36.

Card transactions contributed slightly lower sales around 31 to 32.

```python
#6.Total sales by Time_of_Day
h=a.groupby("Time_of_Day")["money"].sum()
h
```

```python
plt.figure(figsize=(7, 7))
plt.pie(h.values,labels=h.index, autopct="%1.1f%%", startangle=90)
plt.title("Total Sales by Time of Day")
plt.show()
```

```python
#7.Number of transactions by Weekday
i=a.groupby("Weekday")["money"].count()
i
```

```python
plt.figure(figsize=(8, 6))
a["Weekday"].value_counts().sort_index().plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Number of Transactions by Weekday")
plt.xlabel("Weekday")
plt.ylabel("Number of Transactions")
plt.grid(axis="y")
plt.show()
```

```python
#8.Day_of_Month vs Total Sales
j=a.groupby("Day_of_Month")["money"].sum()
j
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(j.index, j.values, color="red", edgecolor="black")
plt.title("Total Sales by Day of Month")
plt.xlabel("Day of Month")
plt.ylabel("Total Sales")
plt.grid(True)
plt.show()
```

```python
fig, axs = plt.subplots(4, 2, figsize=(18, 20))  # 4 rows, 2 columns
axs = axs.flatten()
# Plot 1: Horizontal bar chart (Coffee sales)
axs[0].barh(c.index, c.values)
axs[0].set_title("Total Sales per Coffee Product")
axs[0].set_xlabel("Money")
axs[0].set_ylabel("Coffee Name")
axs[0].grid(axis='x')
# Plot 2: Line chart (Average Sales by Weekday)
axs[1].plot(d.index, d.values, linestyle='-', color="green", marker="o")
axs[1].set_title("Average Sales by Weekday")
axs[1].set_xlabel("Weekday")
axs[1].set_ylabel("Average Sales (money)")
axs[1].grid()
# Plot 3: Bar chart (Total Sales by Hour of Day)
axs[2].bar(e.index, e.values, color="blue", edgecolor="black")
axs[2].set_title("Total Sales by Hour of Day")
axs[2].set_xlabel("Hour of Day")
axs[2].set_ylabel("Total Sales")
axs[2].grid(axis="y")
# Plot 4: Line chart (Average Sales by Month)
axs[3].plot(f.index, f.values, marker="o", color="orange", linewidth=2)
axs[3].set_title("Average Sales by Month")
axs[3].set_xlabel("Month")
axs[3].set_ylabel("Average Sales (money)")
axs[3].grid()
axs[3].tick_params(axis='x', rotation=45)
# Plot 5: Horizontal bar chart (Total Sales by Cash Type)
axs[4].barh(g.index, g.values, color="purple", edgecolor="black")
axs[4].set_title("Total Sales by Cash Type")
axs[4].set_xlabel("Money")
axs[4].set_ylabel("Cash Type")
axs[4].grid(axis="x")
# Plot 6: Pie chart (Total Sales by Time of Day)
axs[5].pie(h.values, labels=h.index, autopct="%1.1f%%", startangle=90)
axs[5].set_title("Total Sales by Time of Day")
# Plot 7: Bar chart (Number of Transactions by Weekday)
axs[6].bar(i.index, i.values, color="skyblue", edgecolor="black")
axs[6].set_title("Number of Transactions by Weekday")
axs[6].set_xlabel("Weekday")
axs[6].set_ylabel("Transactions")
axs[6].grid(axis="y")
# Plot 8: Scatter plot (Total Sales by Day of Month)
axs[7].scatter(j.index, j.values, color="red", edgecolor="black")
axs[7].set_title("Total Sales by Day of Month")
axs[7].set_xlabel("Day")
axs[7].set_ylabel("Total Sales")
axs[7].grid(True)
plt.show()
```
